```python
import pandas as pd
```

```python
df=pd.read_excel('/content/Cu-Ni-Si-alloys.xlsx')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#for numerical columns
df=df.fillna(df.median(numeric_only=True))
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#for categorical columns
for col in df.columns:
  if df[col].dtype=='object':
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
#removing the unnamed columns as they have all missing values
df=df.drop(['Unnamed: 21','Unnamed: 22','Unnamed: 23','Unnamed: 24'],axis=1)
```

```python
df.shape
```

```python
df.info()
```

```python
df.to_excel('/content/cleaned_data.xlsx')
```
